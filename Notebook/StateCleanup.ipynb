State Cleanup Program

The State Cleanup Program manages sites contaminated with hazardous substances or petroleum that aren't on the federal Superfund list. It follows guidelines from the National Oil and Hazardous Substances Pollution Contingency Plan and the Remediation Closure Guide. Most sites are referred to the program by the IDEM Emergency Response Section after a spill or release is reported, but referrals can also come from other IDEM programs, local agencies, or public complaints.

Source: https://www.in.gov/idem/cleanups/investigation-and-cleanup-programs/state-cleanup-program/

```python
df['ZIP'] = df['ZIP'].astype(str)
df_zip['ZIP'] = df_zip['ZIP'].astype(str)
```

```python
df_merged = pd.merge(df, df_zip, on='ZIP', how='inner')
```

In the Cities column, we noticed that some of the cities name were inconsistent (spelling mistakes) for example mentioned below are the different variations of Indianapolis observed in the data.

 'Indianapolis','INDIANAPOLIS','Indianaoplis','Indianpolis','Indianaplois','Indy','INDIANAPOLILS': 'Indianapolis'

```python
#Standardize city names
df_merged['City'] = df_merged['City'].apply(
    lambda x: 'Indianapolis' if x.lower().startswith('ind') else x
)
```

Here, we will be narrowing down to those contaminants which are hazardous but the sites having these are not included in National Priorities List (NPL)
Contaminants are Lead, PCBs, Petroleum.

Source: https://www.in.gov/idem/cleanups/investigation-and-cleanup-programs/state-cleanup-program/

```python
desired_contaminants = ['Lead', 'PCB', 'Petroleum']
df_merged = df_merged[df_merged['Contaminant'].isin(desired_contaminants)]
```

```python
df_final_state_cleanup=df_merged
```
