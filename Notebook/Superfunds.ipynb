Geographic Distribution: <br>
The Superfund sites are spread across various counties in Indiana, including Lake, St. Joseph, Monroe, Madison, Marion, Cass, Bartholomew, Elkhart, Howard, Vigo, Boone, Allen, Owen, Vanderburgh, Kosciusko, Grant, Spencer, Morgan, Hancock, Knox, Jackson, Tippecanoe, Porter, and Whitley counties. This indicates that environmental contamination issues are not confined to one specific area but are distributed throughout the state. <br>

Site Types:<br>
The list includes various types of contaminated sites, such as:<br>
Landfills (e.g., Gary Development Landfill, Northside Sanitary Landfill)<br>
Industrial sites (e.g., Continental Steel Corp., Federated Metals Corp Whiting)<br>
Dump sites (e.g., Fort Wayne Reduction Dump, Marion (Bragg) Dump)<br>
Groundwater contamination plumes (e.g., Broadway Street Corridor Groundwater Contamination, Keystone Corridor Ground Water Contamination)<br>
Chemical plants (e.g., American Chemical Service, Inc., Reilly Tar & Chemical Corp.)<br>
NPL Status:<br>
Most sites are listed as "Final" on the National Priorities List (NPL), indicating they have been formally added to the Superfund program for long-term remedial action. However, some sites are listed as "Deleted," meaning they have been successfully cleaned up and removed from the NPL. One site (Town of Pines Groundwater Plume) is listed as "Non-NPL," suggesting it may be addressed through other environmental programs.<br>
Urban vs. Rural:<br>
The list includes sites in major cities like Indianapolis, Fort Wayne, and South Bend, as well as smaller towns and rural areas, indicating that contamination issues affect both urban and rural communities in Indiana.<br>
Historical Industrial Activity:<br>
Many sites appear to be related to historical industrial activities, such as metal processing, chemical manufacturing, and waste disposal, reflecting Indiana's industrial heritage.<br>
Groundwater Contamination:<br>
There are several entries specifically mentioning groundwater contamination, highlighting the importance of protecting water resources and the long-term effects of certain types of pollution.<br>
Cleanup Progress:<br>
The presence of "Deleted" sites demonstrates that successful cleanups have occurred in Indiana, though many sites remain on the active NPL.<br>
Diverse Contaminants:<br>
