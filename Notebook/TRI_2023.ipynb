```python
print(df.info())

# Display summary statistics
print(df.describe())

# Check for missing values
print(df.isnull().sum())
```

```python
# Analyze categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    print(f"\nUnique values in {col}:")
    print(df[col].value_counts().head())

# Visualize top industries
plt.figure(figsize=(12,6))
df['23. INDUSTRY SECTOR'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Industry Sectors')
plt.xlabel('Industry Sector')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_industries.png')
plt.close()
```

```python
# Plot facility locations
plt.figure(figsize=(10,8))
plt.scatter(df['13. LONGITUDE'], df['12. LATITUDE'], alpha=0.5)
plt.title('Geographical Distribution of Facilities')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.tight_layout()
plt.savefig('facility_locations.png')
plt.close()

# Analyze distribution of facilities by county
plt.figure(figsize=(12,6))
df['7. COUNTY'].value_counts().head(20).plot(kind='bar')
plt.title('Top 20 Counties by Number of Facilities')
plt.xlabel('County')
plt.ylabel('Number of Facilities')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_counties.png')
plt.close()
```

```python
# Select numerical columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Analyze release-related columns
release_cols = [col for col in numeric_cols if 'RELEASE' in col.upper()]

# Calculate total releases
df['TOTAL_RELEASES'] = df[release_cols].sum(axis=1)

# Plot distribution of total releases
plt.figure(figsize=(10,6))
sns.histplot(df['TOTAL_RELEASES'], kde=True, bins=50)
plt.title('Distribution of Total Releases')
plt.xlabel('Total Releases')
plt.xscale('log')
plt.tight_layout()
plt.savefig('total_releases_distribution.png')
plt.close()

# Top facilities by total releases
top_facilities = df.nlargest(10, 'TOTAL_RELEASES')
plt.figure(figsize=(12,6))
plt.bar(top_facilities['4. FACILITY NAME'], top_facilities['TOTAL_RELEASES'])
plt.title('Top 10 Facilities by Total Releases')
plt.xlabel('Facility Name')
plt.ylabel('Total Releases')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_facilities_releases.png')
plt.close()
```

```python
# Analyze most common chemicals
plt.figure(figsize=(12,6))
df['37. CHEMICAL'].value_counts().head(20).plot(kind='bar')
plt.title('Top 20 Most Reported Chemicals')
plt.xlabel('Chemical')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_chemicals.png')
plt.close()

# Analyze carcinogenic chemicals
carcinogen_counts = df['46. CARCINOGEN'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(carcinogen_counts, labels=carcinogen_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Carcinogenic Chemicals')
plt.tight_layout()
plt.savefig('carcinogen_distribution.png')
plt.close()
```

```python
# Correlation analysis of release-related columns
release_correlation = df[release_cols].corr()
plt.figure(figsize=(12,10))
sns.heatmap(release_correlation, cmap='coolwarm', annot=False)
plt.title('Correlation of Release-Related Columns')
plt.tight_layout()
plt.savefig('release_correlation.png')
plt.close()
plt.show()
```

```python
# Analyze trends over time (if multiple years are present)
if df['1. YEAR'].nunique() > 1:
    yearly_releases = df.groupby('1. YEAR')['TOTAL_RELEASES'].sum()
    plt.figure(figsize=(10,6))
    yearly_releases.plot(kind='line', marker='o')
    plt.title('Total Releases Over Time')
    plt.xlabel('Year')
    plt.ylabel('Total Releases')
    plt.tight_layout()
    plt.show()

print("\nSummary of Findings:")
print(f"1. Total number of facilities: {len(df)}")
print(f"2. Number of unique chemicals reported: {df['37. CHEMICAL'].nunique()}")
print(f"3. Top industry sector: {df['23. INDUSTRY SECTOR'].value_counts().index[0]}")
print(f"4. County with most facilities: {df['7. COUNTY'].value_counts().index[0]}")
print(f"5. Percentage of carcinogenic chemicals: {carcinogen_counts['YES'] / len(df) * 100:.2f}%")
print(f"6. Facility with highest total releases: {top_facilities['4. FACILITY NAME'].iloc[0]}")
print(f"7. Most commonly reported chemical: {df['37. CHEMICAL'].value_counts().index[0]}")
```
